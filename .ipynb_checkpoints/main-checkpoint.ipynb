### 1.1 Import necessary libraries

````python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, log_loss, r2_score, plot_confusion_matrix
from imblearn.over_sampling import SMOTE
````

### 1.2 Check first 5 rows
We can already see there are some ```NaN``` values. We can address these values later.

````python
df = pd.read_csv('water_potability.csv')
df.head()
````

### 1.3 Statistical summary of each applicable column
We see that columns ```ph```, ```Sulfate```, and ```Trihalomethanes``` have missing values.

## 2. Clean Data
1. Check which columns have NaN values
2. Fill the NaN values with the mean of each applicable feature/column

### 2.1 Check which columns have NaN values

````python
nan_rows = df[df.isna().any(axis=1)]
nan_rows.head()
````

````python
nan_cols = df.columns[df.isna().any()].tolist()
nan_cols
````

### 2.2 Fill the NaN values with the mean of each applicable feature/column

````python
for col in nan_cols:
    df[col].fillna(df[col].mean(), inplace = True)
df.head()
````

## 3. Transform Data
1. Use Sklearn's ```StandardScaler``` class to standardize the dataset (all columns except ```Potability```)

### 3.1 Use Sklearn's ```StandardScaler``` class to standardize the dataset (all columns except ```Potability```)

````python
scaler = StandardScaler()
for col in df.columns[:-1]:
    df[col] = scaler.fit_transform(np.array(df[col]).reshape(-1,1))
df.head()
````

## 4. Define and Train Models
1. Create ```X``` and ```y``` datasets
2. Create training and testing datasets
3. K-Nearest Neighbors
4. Linear SVM
5. Gaussian Process
6. Decision Tree
7. Random Forest
8. Neural Network
9. AdaBoost
10. Naive Bayes
11. QDA

````python
models = {}
````

### 4.1 Create ```X``` and ```y``` datasets

````python
X = df[df.columns[:-1]]
y = df['Potability']
print(X.head(), y.head())
````

````python
sm = SMOTE(random_state = 42)
X, y = sm.fit_resample(X, y)
print(X.head(), y.head())
````

### 4.2 Create Training and Testing Datasets

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
````

````python
y_train.value_counts()
````

### 4.3 K-Nearest Neighbors

````python
ks = 15
means = np.array([])
for i in range(35):
    scores = np.array([])
    for k in range(1, ks):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
        knn = KNeighborsClassifier(n_neighbors = k)
        knn.fit(X_train, y_train)
        scores = np.append(scores, knn.score(X_test, y_test))
    means = np.append(means, np.where(scores == np.amax(scores))[0][0] + 1)
    plt.plot(range(1, ks), scores, alpha=.5)
plt.title('Accuracy Score for K up to 15')

plt.xlabel('K value')
plt.ylabel('Accuracy')

plt.xticks(range(1, ks), [i for i in range(1, ks)])
plt.grid()
plt.show()
````

````python
print(f'Most common accurate value of K: {int(stats.mode(means)[0][0])}')
````

````python
models['KNN'] = KNeighborsClassifier(n_neighbors = 1)
models['KNN'].fit(X_train, y_train)
models['KNN'].score(X_test, y_test)
````

### 4.4 Linear SVM

````python
models['SVM'] = SVC()
models['SVM'].fit(X_train, y_train)
models['SVM'].score(X_test, y_test)
````

### 4.5 Gaussian Process

````python
models['Gaussian Process'] = GaussianProcessClassifier(random_state = 42)
models['Gaussian Process'].fit(X_train, y_train)
models['Gaussian Process'].score(X_test, y_test)
````

### 4.6 Decision Tree

````python
models['Decision Tree'] = DecisionTreeClassifier(criterion = 'entropy', random_state = 42)
models['Decision Tree'].fit(X_train, y_train)
print(models['Decision Tree'].score(X_test, y_test))
plot_tree(models['Decision Tree'], max_depth = 8, filled = True, rounded = True)
plt.show()
````

### 4.7 Random Forest

````python
models['Random Forest'] = RandomForestClassifier(criterion = 'entropy', random_state = 42)
models['Random Forest'].fit(X_train, y_train)
models['Random Forest'].score(X_test, y_test)
````

### 4.8 Neural Network

````python
models['MLP'] = MLPClassifier(max_iter = 1500, random_state = 42)
models['MLP'].fit(X_train, y_train)
models['MLP'].score(X_test, y_test)
````

### 4.9 AdaBoost

````python
models['Ada'] = AdaBoostClassifier(n_estimators = 150, random_state = 42)
models['Ada'].fit(X_train, y_train)
models['Ada'].score(X_test, y_test)
````

### 4.10 Naive Bayes

````python
models['Naive Bayes'] = GaussianNB()
models['Naive Bayes'].fit(X_train, y_train)
models['Naive Bayes'].score(X_test, y_test)
````

### 4.11 QDA

````python
models['QDA'] = QuadraticDiscriminantAnalysis()
models['QDA'].fit(X_train, y_train)
models['QDA'].score(X_test, y_test)
````

## 5. Measure Models' Accuracy

````python
metrics = {
    'Confusion Matrix': confusion_matrix,
    'Accuracy': accuracy_score,
    'F1': f1_score,
    'Recall': recall_score,
    'Precision': precision_score,
    'Log Loss': log_loss,
}

def score(y_true, y_pred):
    d = {}
    for metric, f in metrics.items():
        d[metric] = f(y_true, y_pred)
    return d
````

````python
model_scores = {}
for model, c in models.items():
    model_scores[model] = score(y_test, c.predict(X_test))  
print(model_scores)
````

````python
colors = ['lightcoral', 'limegreen', 'skyblue', 'orange']
norm_labels = np.array(['Accuracy', 'F1', 'Recall', 'Precision'])
ind = np.arange(9)
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(5)
for i, scores in enumerate(model_scores.values()):
    sep = -.3
    color_id = 0
    label_score = np.array([])
    for score in norm_labels:
        label_score = np.append(label_score, scores[score])
        plt.bar(ind[i]+sep, scores[score], width=0.2, color=colors[color_id], align='center', edgecolor='black')
        sep += .2
        color_id += 1
    print(f'Overall accuracy (mean) of {list(model_scores.keys())[i]}: {np.average(label_score)}')
        
plt.title('Accuracy Scores per Model')
plt.gcf().set_dpi(144)  
plt.ylabel('Score')
plt.xlabel('Model Used')
        
plt.legend(norm_labels)
plt.xticks(ind, models.keys(), rotation=20)
plt.show()
````

````python
for model, func in models.items():
    disp = plot_confusion_matrix(func, X_test, y_test)
    disp.ax_.set_title(model)
````

## 6. Conclusions
1. **Most** accurate model (score mean): K-Nearest Neighbors with ```K=1``` (~0.75)
2. **Least** accurate model (score mean): Naive Bayes (~0.51)
